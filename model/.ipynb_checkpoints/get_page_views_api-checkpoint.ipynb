```python
import requests
import pandas as pd 
import json
```

```python
endpoint = "https://wikimedia.org/api/rest_v1"
```

```python
def return_params(article_name):
    project = "en.wikipedia.org"
    access = "all-access"
    agent = "all-agents"
    article = article_name
    granularity = "monthly"
    start = "20010101"
    end = "20170101"
    params = "/metrics/pageviews/per-article/{}/{}/{}/{}/{}/{}/{}".format(project,access,
    agent, article, granularity, start, end)
    return params
```
