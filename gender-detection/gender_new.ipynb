```python
def print_stats(df, col):
    """ Prints the number of instances in the data set, number of instances with a known value,
        absolute and relative frequencies.
    """
    dic = df[col].value_counts().to_dict()
    total = sum(dic.values())
    print("Total in dataset: ", len(df))
    print("Total gender determined: ", total)
    print("Ratio known/total: ", total/len(df))
#     print(sum(dic.values()))
    print(dic)
#     print 
#     print(df[col].value_counts())
    dic.update((k, v / total ) for k,v in dic.items())
    print(dic)
    
```

```python
def add_name_col(df):
    df["first_name"] = df.apply(get_f_name, axis=1)
    return df
```

```python
def crop_name(name_list):
    if name_list == []:
        return "unknown"
    else: 
        if name_list[0].split(" ")[1] == "sir" or name_list[0].split(" ")[1] == "dr" or name_list[0].split(" ")[1] == "the":
            return name_list[0].split(" ")[2]
        return name_list[0].split(" ")[1]
```

```python
def assign_gender(name, dic):
    if name in dic:
        return dic[name]
    else:
        return None
```

```python
def get_f_name(x):
    """ Returns first name based on the WikiURL and name column values
    """
    url_name = name_from_url(x['WikiURL'])
    list_name = crop_name(x['name'])
#     print("LIST:"+list_name+"URL:"+url_name)
    if len(url_name) == 1 or "-" in url_name:
#         print("LIST")
        return list_name
#     print("URL")
    return url_name
```

```python
def name_from_url(url):
    """ Extracts first name based on WikiURL values
    """
    name = url.split("/")[-1].split("_")[0].strip(",").strip(".").lower()
    if name == "sir" or name == "the" or name == "dr":
        name = url.split("/")[-1].split("_")[1].strip(",").strip(".").lower()
    if name in ["sir","the",["dr"]]:
        return "unknown"
    else: 
        return name
```

```python
def get_full_name(url):
    """ Returns full name extracted from wikipedia url in plus separated form Nikola+Tesla
    """
    name_parts = url.split("/")[-1].split("_")
    return "+".join(name_parts)
```

```python
def final_gender(gnd_col, fpp_col):
    """ Checks gender assigned by genderize and face ++ and assigns it to the gender col
    """
    if gnd_col != gnd_col and fpp_col != fpp_col:
        return 'unknown'
    elif gnd_col == 'male' or 'female':
        return gnd_col
    elif fpp_col == 'male' or 'female':
        return fpp_col
```
